```python
# Building Collaborative filtering model from scratch
"""
We will recommend movies based on user similarity and item-item similarity.
For that, first we need to calculate the number of unique users and movies
"""
n_users = ratings.user_id.unique().shape[0]
n_items = ratings.movie_id.unique().shape[0]
```

```python
"""
Now, we will create a user-item matricx which can be used to calculate the
similarity between users and items
"""
data_matrix = np.zeros((n_users, n_items))
for line in ratings.itertuples():
    data_matrix[line[1]-1, line[2]-1] = line[3]
```

```python
"""
Each row in data_matrix corresponds to users and each column corresponds to
each movie ratings
"""
"""
We now will calculate the similarity. We can use the pairwise_distance function
from sklearn to calculate the cosine similarity
"""

from sklearn.metrics.pairwise import pairwise_distances
user_similarity = pairwise_distances(data_matrix, metric='cosine')
item_similarity = pairwise_distances(data_matrix.T, metric='cosine')

```

```python
"""
The above calculation is computationally expensive and would work bad for 
large dataset

The pairwise_distance function calculates the distance between corresponding
row elements and therefore for calculating the similarity score for items we
took transpose of the data_matrix
"""
```

```python
"""
The above has given us the item-item and user-user similarity in an array form.
The next step is to make predictions based on these similarities.
"""

def predict(ratings, similarity, type='user'):
    if type == 'user':
        mean_user_rating = ratings.mean(axis=1)
        
        # We use np.newaxis so that mean_user_rating has some format as rating
        ratings_diff = (ratings - mean_user_rating[:, np.newaxis])
        
        pred = mean_user_rating[:, np.newaxis] + similarity.dot(ratings_diff) / np.array([np.abs(similarity).sum(axis=1)]).T
        
    elif type == 'item':
        pred = ratings.dot(similarity) / np.array([np.abs(similarity).sum(axis=1)])
            
    
    return pred
```

```python
"""
Finally we will make predictions based on user similarity and item similarity
"""

user_prediction = predict(data_matrix, user_similarity, type='user')
item_prediction = predict(data_matrix, item_similarity, type='item')
```
